```python
#importing libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("austin_weather_final.csv")
data
```

```python
X = data.drop(['PrecipitationSumInches'], axis = 1 )
```

```python
Y = data["PrecipitationSumInches"]
```

```python
#reshaping it into 2d vector
Y = Y.values.reshape(-1,1)
```

```python
day_index = 798
days = [i for i in range (Y.size)]
days
```

```python
#initiolie the linear classifier

clf = LinearRegression()

#train classifier
clf.fit(X,Y)
```

```python
#plot a graph

print("the Precipitation trend graph: ")
plt.scatter(days, Y, color="Green")
plt.scatter(days[day_index], Y[day_index], color = "Red")
plt.title("precipitation trend")
plt.xlabel("Days")
plt.ylabel("precipitation in inches ")
plt.show()
x_vis = X.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureLowInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
```

```python
print("the Precipitation trend graph: ")

for i in range(x_vis.columns.size):
  plt.subplot(3,2,i+1)
  plt.scatter(days, x_vis[x_vis.columns.values[i]],  color = "Green")
  plt.scatter(days[day_index], x_vis[x_vis.columns.values[i]][day_index], color = "red")
  plt.title(x_vis.columns.values[i])
plt.show()
```
