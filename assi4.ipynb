```python
import numpy as np
array_1d = np.arange(1, 21)

#a. calculate sum, mean, median, standard deviation
sum_ = np.sum(array_1d)
mean_ = np.mean(array_1d)
median_ = np.median(array_1d)
std_dev_ = np.std(array_1d)

print("array:", array_1d)
print("sum:", sum_)
print("mean:", mean_)
print("median:", median_)
print("Standard deviation:", std_dev_)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]


plt.plot(x, y, marker='o', linestyle='-', color='blue')

plt.title("Simple Line Plot")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid(True)
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]


plt.bar(students, marks, color='skyblue')

plt.title("Marks Scored by Students")

plt.xlabel("Students")
plt.ylabel("Marks")

plt.ylim(0, 100)
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# Explode the largest revenue region
explode = [0.1 if r == max(revenue) else 0 for r in revenue]

# Pie chart
plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140)

plt.title("Company Revenue by Region", color = 'pink')

plt.axis('equal')
plt.show()
```
