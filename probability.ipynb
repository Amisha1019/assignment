```python
import numpy as np

values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]

sample = np.random.choice(values, size=1000, p=probabilities)
print(f"Empirical Mean: {np.mean(sample):.4f}")
print(f"Empirical Variance: {np.var(sample):.4f}")
print(f"Empirical Std Dev: {np.std(sample):.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Generate 10,000 values from uniform distribution
uniform_data = np.random.uniform(0, 1, 10000)

# Sampling and calculating means
sample_means = []
for _ in range(1000):
    sample = np.random.choice(uniform_data, size=30)
    sample_means.append(np.mean(sample))

# Plotting
plt.figure(figsize=(12,5))

# Original Uniform Distribution
plt.subplot(1,2,1)
sns.histplot(uniform_data, bins=30, kde=False, color='lightgreen')
plt.title("Original Uniform Distribution")

# Distribution of Sample Means
plt.subplot(1,2,2)
sns.histplot(sample_means, bins=30, kde=True, color='salmon')
plt.title("Distribution of Sample Means (CLT)")

plt.tight_layout()
plt.show()
```
