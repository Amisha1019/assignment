```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
 dataset.head()
```

```python
corona_dataset_aggregated = dataset.groupby("Country/Region").sum(numeric_only=True)
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
corona_dataset_aggregated.loc['China'].plot()
corona_dataset_aggregated.loc['India'].plot()
corona_dataset_aggregated.loc['Spain'].plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc['China'][:3].plot()
plt.xlabel("Date")
plt.ylabel("Number of Confirmed Cases")
plt.title("COVID-19 Confirmed Cases in China")
plt.show()
```

```python
corona_dataset_aggregated.loc['China'].diff().plot()
```

```python
corona_dataset_aggregated.loc['India'].diff().max()
```

```python
corona_dataset_aggregated.loc['Spain'].diff().max()
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for country in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[country].diff().max())
corona_dataset_aggregated['Max Infection Rate'] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset = pd.DataFrame(corona_dataset_aggregated['Max Infection Rate'])
```
